**Python Assignment Operators**
Screenshot 2026-03-10 at 12.34.16 PM.png

```python
#Assignment operator
count = 0
count = count + 1

print(count)

#Compound Assignment Operators
count = 0
count += 1

print(count)

#Add and Assign (+=)
count = 0
count += 1  # Same as: count = count + 1

print(count)

#Subtract and Assign (-=)
quantity = 5
quantity -= 2  # Same as: quantity = quantity- 2

print(quantity)

#Multiply and Assign (*=)
a = 10
a *= 2  # Same as: a = a * 2

print(a)  # Output: 20

#Divide and Assign (/=)
amount = 10
amount //= 3  # Same as: amount = amount // 3

print(amount)

#Modulus and Assign (%=)
amount = 10
amount %= 3

print(amount)

#Exponentiate and Assign (**=)
a = 2
a **= 3  # Same as: a = a ** 3
print(a)
```

**Python Comparison Operators** : Python has six comparison operators to compare both **numbers** and **strings**

Less than ( < )

Less than or equal to (<=)

Greater than (>)

Greater than or equal to (>=)

Equal to ( == )

Not equal to ( != )

```python
'''The expression 'apple' < 'orange' returns True because the letter a in apple is before the letter o in orange.
the 'banana' < 'apple' returns False because the letter 'b' is after the letter 'a'.'''

result = 'apple' < 'orange'
print(result)

result = 'banana' < 'apple'
print(result)
```

```python
#Less than or equal to operator (<=)
result = 20 <= 20
print(result)

result = 10 <= 20
print(result)

result = 30 <= 30
print(result)

#Equal To operator (==)
result = 20 == 10
print(result)

result = 20 == 20
print(result)

#Not Equal To operator (!=)
result = 20 != 20
print(result)

result = 20 != 10
print(result)

result = 'apple' != 'apple'
print(result)

result = 'apple' != 'orange'
print(result)
```
